Standard Normal Distribution

```python
mu = 0
sigma = 1

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, mu, sigma)
# Plot the Gauss-curve
plt.plot(dist_x, dist_y)
# Show the mean with an orange line
plt.axvline(mu, color="orange", lw=2)
plt.ylim(0, 0.5)
```

One way to compare different normal populations is to standardise the scores - by looking at their distance from the mean, then dividing by the size of the standard deviation.  
To find the standardised score, $z$, from a raw score, $x$, use the conversion   
$z=(x - μ)/\sigma$  
where $\mu$ is the mean and $\sigma$ is the standard deviation of the raw scores.  
The standard score is often called the $z-score$

```python
# Probability density at x
# norm.pdf(x, loc=m, scale=s)

mu = 5
sigma = 1.5
x = 6

pdf_value = stats.norm.pdf(x, loc = mu, scale = sigma)
print(f"pdf {pdf_value}")

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, mu, sigma)
# Plot the Gauss-curve
plt.plot(dist_x, dist_y)
plt.ylim(0, 0.5)
# Show x with a green line
plt.axvline(x, color="green");
```

```python
# Left-tail probability 𝑃(𝑋 < x)
# cdf = cumulative distribution function
# norm.cdf(x, loc=m, scale=s) 

mu = 5
sigma = 1.5
x = 6

cdf_value = stats.norm.cdf(x, loc = mu, scale = sigma)
print(f"cdf {cdf_value}")

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, mu, sigma)
# Plot the Gauss-curve
plt.plot(dist_x, dist_y)
plt.ylim(0, 0.5)

# Fill the area left of x
plt.fill_between(dist_x, 0, dist_y, where=dist_x <= x, color='lightblue')

# Show x with a green line
plt.axvline(x, color="green");
```

```python
# Right-tail probability 𝑃(𝑋 > x)
# sf = survival function
# norm.sf(x, loc=m, scale=s)

mu = 5
sigma = 1.5
x = 6

sf_value = stats.norm.sf(x, loc = mu, scale = sigma)
print(f"sf {sf_value}")

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, mu, sigma)
# Plot the Gauss-curve
plt.plot(dist_x, dist_y)
plt.ylim(0, 0.5)

# Fill the area left of x
plt.fill_between(dist_x, 0, dist_y, where=dist_x >= x, color='lightblue')

# Show x with a green line
plt.axvline(x, color="green");
```

```python
# p% of observations are expected to be lower than result
# inverse survival function
# norm.isf(1-p, loc=m, scale=s) 

mu = 5
sigma = 1.5
p = 0.4 # percentage!

isf_value = stats.norm.isf(p, loc = mu, scale = sigma)
print(f"isf {isf_value}")

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, mu, sigma)
# Plot the Gauss-curve
plt.plot(dist_x, dist_y)
plt.ylim(0, 0.5)

# Fill the area left of x
plt.fill_between(dist_x, 0, dist_y, where=dist_x <= isf_value, color='lightblue')

# Show x with a green line
plt.axvline(isf_value, color="green");
```

##The Central Limit Theorem
If the size of the sample is sufficiently large, the probability
distribution of the sample mean will approximate a normal
distribution, regardless of the probability distribution of the
underlying population.


```python
sns.displot(x = mpg['acceleration'], bins=20);
```

##Confidence interval - Large sample

**True Meaning of a 95% Confidence
Interval**  
Given the data in a particular sample, a 95% confidence
interval for the mean will either include the
(unknown) population mean or it won't. The true
meaning of a 95% confidence interval is that if the
same procedure is used on many different random
samples, about 95% of the resulting confidence intervals
will include the population mean, and only about
5% won’t. Therefore, you can be 95% confident that
any particular confidence interval is a "good" one.


```python
population = np.arange(1,1000)
mu = arr.mean()
sigma = arr.std()

sample_size = 50            # the size of the sample
num_samples = 10000  

intervalOK = 0
intervalNotOK = 0

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    lowerlimit = sample.mean() - (1.96 * sigma / math.sqrt(sample_size))
    upperlimit = sample.mean() + (1.96 * sigma / math.sqrt(sample_size))  
    if (lowerlimit <= mu) and (mu <= upperlimit):
      intervalOK += 1
    else:
      intervalNotOK += 1

print(f"IntervalOK = {intervalOK / num_samples} - IntervalNotOK = {intervalNotOK / num_samples}")
```

##Confidence Interval - Small sample  
For a small sample the central limit theorem is no longer valid.  
Instead we can say:  
If a population 𝑋 has a normal distribution (𝑋 ∼ 𝑁𝑜𝑟(𝜇, 𝜎)) and you take a
small sample with mean 𝑥 and standard deviation 𝑠, then  
$t = (\overline x$ − 𝜇) / (𝑠 / √𝑛)  
will behave as a $t$-distribution with 𝑛 − 1 degrees of freedom

```python
mu = 0
sigma = 1
x = 0.5

# X-values
dist_x = np.linspace(mu - 4 * sigma, mu + 4 * sigma)
# Y-values for drawing the Gauss curve
dist_y = stats.norm.pdf(dist_x, loc = mu, scale = sigma)
# Y-values for drawing the t curve
dist_y_t = stats.t.pdf(dist_x, 4)
# Plot the Gauss-curve and the t-curve
plt.plot(dist_x, dist_y, color='blue')
plt.plot(dist_x, dist_y_t, color='orange')
plt.ylim(0, 0.5)
```
