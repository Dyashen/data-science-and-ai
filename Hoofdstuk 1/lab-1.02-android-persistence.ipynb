Je hebt enkel Pandas nodig. Importeer de dataset. Bekijk het CSV-bestand zeker nog eens om te kijken hoe alle waarden gescheiden zijn. In dit geval worden alle waarden gescheiden door een puntkomma. Geef dit mee als 'delimiter' parameter.

```python
import pandas as pd
ap = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-en-labs/main/data/android_persistence_cpu.csv', delimiter=';')
ap
```

1. Print het aantal rijen en kolommen uit.
2. Verander de kolommen met een object-type naar het meest passende type.
3. Geef info over de gehele dataset.
4. Toon alle datatypes van de dataset.

```python
# Print het aantal rijen en kolommen uit.
print(f'De dataset bevat {ap.shape[0]} rijen & {ap.shape[1]} kolommen ')

# PType en DataSize zijn twee kwalitatieve waarden die nu worden gezien als een 'object'. Vorm de twee kolommen om naar een categorietype. 
ap['PersistenceType'] = ap['PersistenceType'].astype('category')
ap['DataSize'] = ap['DataSize'].astype('category')

# Geef info over de dataset.
ap.info()

# Toon alle datatypes van de dataset.
ap.dtypes

```

1. Geef een omschrijving van iedere variabele in de dataset.
2. Bereken het gemiddelde van de Time-variabele. Hou rekening met niet-ingevulde velden.

```python
# Toon de statistieken van iedere kolom. Hiervoor moet je iedere kolom 
ap['PersistenceType'].describe()
ap['Time'].describe()
ap['DataSize'].describe()


# Extra:
# Bereken het gemiddelde van de Time-variabele. Zorg ervoor dat ieder niet-ingevuld element wordt vervangen door het gemiddelde.
m = ap.Time.mean()
ap['Time'] = ap.Time.fillna(m)
```

1. Haal de unieke velden op van iedere categorische variabele. 
2. Vraag een frequentietabel op van beide variabelen.

```python
# Unieke waarden
ap['PersistenceType'].unique()
ap['DataSize'].unique()

# Frequentietabel opvragen
ap['PersistenceType'].value_counts()
ap['DataSize'].value_counts()
```

Opmerking: Crosstab komt pas in hoofdstuk 4 aan bod.

Met een kruistabel of contingency table toon je een mogelijk verband tussen een kwalitatieve en een kwantitatieve variabele. 
In dit geval is PersistenceType een categorische variabele. DataSize maakt gebruik van getallen en is hier de kwantitatieve variabele.
Crosstab verwacht twee parameters in dezelfde volgorde: de afhankelijke variabele en een onafhankelijke variabele.

```python
pd.crosstab(ap['PersistenceType'], ap['DataSize'])
```
