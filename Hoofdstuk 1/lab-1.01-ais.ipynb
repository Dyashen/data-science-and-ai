De eerste (vijf) rijen van een dataframe tonen doe je aan de hand van head(). De laatste doe je met tail(). Je kan het aantal rijen specifiëren door dit mee te geven als parameter.

```python
ais = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-en-labs/main/data/ais.csv')
ais.head()
ais.tail() 
```

Toon algemene informatie over de dataset.

* Shape geeft je een tupel terug van twee waarden: het aantal rijen en het aantal kolommen.
* Dtypes geeft je de datatype van ieder variabele terug.
* Describe gaat algemene informatie geven over ieder variabele.
* Value_counts telt het aantal voorkomens van waarden in een dataset. Met deze functie kan je een frequentietabel opbouwen.

```python
rijen = ais.shape[0]
kolommen = ais.shape[1]
print(f'Deze dataset bestaat uit {rijen} rijen en {kolommen} kolommen.')

typeVar = ais.dtypes

algemeneInfo = ais.describe

# Het aantal kolommen van ieder type opvragen.
# Vraag eerst op wat de datatypes zijn van iedere kolom met dtypes
# Bereken hierop het aantal voorkomens met value_counts
ais.dtypes.value_counts()
```

ID is de index van de dataset. Dit moeten wij nog instellen door middel van .set_index(..naam van de kolom..)
Let op! Deze methode is immutable en past de dataset niet aan. Spreek de methode aan en sla de dataset op zoals hieronder.

```python
ais = ais.set_index('id')
```

Sommige kwalitatieve variabelen worden nog opgeslaan onder een 'object' type. Dit mag niet en we moeten dit aanpassen naar een categorische variabele. Om een kolom op te halen zijn er twee mogelijkheden. Ofwel doe je .kolomnaam ofwel gebruik je vierkante haakjes met daartussen de kolomnaam.

Een datatype omzetten doe je met 'astype'. Dit is een immutable methode dus sla deze nieuwe dataset op of overschrijf de huidige dataset zoals hieronder.

```python
ais.sex = ais.sex.astype('category')

# Alternatieve manier.
ais['sex'] = ais['sex'].astype('category')
```

## Oefeningen

```python
# Tweede rij van de dataset
# OPM: gebruik [[...]] voor een dataframe en [...] voor een serie
ais.iloc[[2]]

# Rij vier tot en met zes
# bouw eerst een filter op --> ais['id'].isin([4,5,6])
# vraag dan alles van de dataset op die voldoet aan de filter --> ais[filter]
ais[ais['id'].isin([4,5,6])]

# Kolom zes tot en met acht.
# gebruik .loc[:, ['kolom1', 'kolom2', ... ]] om een specifiek aantal kolommen op te vragen
# voor één kolom kan je het houden bij ais['kolom'] of ais.kolom
ais.loc[:, ['ferr', 'bmi', 'ssf']]

# De variabele pcBfat
# Verschillende manieren om een kolom op te halen:
ais['pcBfat']
ais.pcBfat
ais.loc[:, 'pcBfat']

# Alles voor de sport 'netball'
# maak eerst een filter op --> ais['sport'] == 'Netball'
ais[ais['sport'] == 'Netball']


# Enkel de kolom 'wt' voor netball-spelers
# ais['kolom'][filter]
ais['wt'][ais['sport'] == 'Netball']

# Alternatieve manier. Hier kan je efficiënt de filters aanpassen in grote python documenten.
kolom = 'wt'
filter = ais['sport'] == 'Netball'
ais[kolom][filter]

# Welke atleten hebben een bmi groter dan 26?
# voorzie een lijst van alle unieke waarden
# voorzie een frequentietabel van hoe vaak een sport voorkomt
filter = ais['bmi'] > 26
kolom = 'sport'

ais[kolom][filter].unique # alle unieke kolommen
ais[kolom][filter].value_counts() # frequentietabel
```
